```python
import torch
```

# **Primeiros passos com tensores**

### Criação de tensores

##### - De listas para Tensores


```python
# Criar tensores a partir de listas
lista = [[1,2,3],
         [4,5, 6]]

tns = torch.Tensor(lista)
print(tns.dtype)
print(tns) # De padrão vem float

tns2 = torch.LongTensor(lista)
print(tns2.dtype)
print(tns2) # Agora vem inteiro
```

##### - De arrays para tensores

```python
# Criar tensores a partir de array numpy
import numpy as np
arr = np.random.rand(3,4) #3 linhas por 4 colunas
print(arr)

tns = torch.from_numpy(arr)
print(tns) #Preserva o tipo do array
```

##### - Inicialização de tensores do zero

```python
#Já iniciar como tensor
tns1 = torch.ones(2,3)
tns0 = torch.zeros(4,5)
tnsr = torch.randn(3,3) #Distribuição normal aleatoria

print(tns1)
print(tns0)
print(tnsr)
```

##### - De tensores para array

```python
print(tnsr)
print(tnsr.dtype)
arr = tnsr.data.numpy()
print(arr)
print(arr.dtype)
```

### Indexação (Acessar um valor especifico)

##### - Substituição

```python
print(tnsr)
tnsr[0,2] = 100
print(tnsr)
```

##### - Pegar valores especificos

```python
print(tnsr)

print("")
print(tnsr[:2]) #Pegar as duas primeiras linhas
print("")
print(tnsr[:2, 2]) #Pegar a terceira coluna das duas primeiras linhas
```

### Operações com tensores

```python
tns_soma = torch.randint(low=0, high=10, size=(3, 3))
print(tns_soma)
print("")
tns_soma2 = torch.randint(low=0, high=10, size=(3, 3))
print(tns_soma2)
print("")
tns_soma3 = tns_soma + tns_soma2 #é preciso ter as mesmas dimensões
print(tns_soma3)
```

### Função Size() e View()

```python
print(tnsr.size())
```

```python
# 2 linhas, 3 colunas, 2 de profundidade (cubo)
tns = torch.randn(2, 2, 3)
print(tns)
print("")
#Isso resultado em um tensor com 12 elementos

tns_achatado = tns.view(12) # Achatamos em uma unica linha e coluna
print(tns_achatado)
print("")
tns_achatado = tns.view(2, 6) # Achatamos em duas linhas e 6 colunas
print(tns_achatado)
print("")

```
