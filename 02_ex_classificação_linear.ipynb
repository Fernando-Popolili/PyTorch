```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(46)

X, Y = make_classification(n_features = 2, n_redundant = 0, n_informative = 1, n_clusters_per_class = 1)

plt.scatter(X[:, 0], X[:, 1], marker = 'o', c = Y, s = 25, edgecolor = 'k')
```

```python
print(X.shape)
```

```python
def plotmodel(w1,w2,b):

  plt.scatter(X[:, 0], X[:, 1], marker = 'o', c = Y, s = 25, edgecolor = 'k')

  xmin, xmax = plt.xlim()
  ymin, ymax = plt.ylim()
  plt.xlim(xmin, xmax)
  plt.ylim(ymin, ymax)

  x = np.linspace(-2, 4, 50)
  y = (-w1*x - b)/w2
  plt.axhline(0, color='black', linestyle='--')
  plt.axvline(0, color='black', linestyle='--')
  plt.plot(x, y)
  plt.grid(True)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Classificação Linear')


# w1 = -3
# w2 = 5
# b = 3

#mudar até encontrar a reta que melhor separa
w1 = 5
w2 = 1
b = 0.4

plotmodel(w1,w2,b)
```

```python
def classify(ponto, w1, w2, b):
  ret1 = w1*ponto[0] + w2*ponto[1] + b
  if ret1 >= 0:
    return 1, "Amarelo"
  else:
    return 0, "Azul"


p = (2, -1)

classe, cor = classify(p, w1, w2, b)

print(f"O ponto {p} é da classe {cor}.")

plotmodel( w1, w2, b)
plt.plot(p[0], p[1], marker='^', color = 'black', markersize = 25)
```

```python
acertos = 0
for x in range(len(X)):
  categ, _ = classify(X[x], w1, w2, b)
  if categ == Y[x]:
    acertos += 1

print("Acurracia:", acertos)

#86% de acuracia
```
