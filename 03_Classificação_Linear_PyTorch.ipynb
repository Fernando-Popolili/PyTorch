```python
import torch
from torch import nn # Modulo de redes neurais
import numpy
```

```python
torch.manual_seed(42)
#Instanciando o perceptron - gerando os pesos e o bias aleatoriamente
perceptron = nn.Linear(in_features=3, out_features=1)

# Versão resumida nn.Linear(3, 1)
print(perceptron)
```

```python
#Acessando esses valores
for name, tensor in perceptron.named_parameters():
    print(name, '\n', tensor.data, '\n')
```

```python
# Salvando os valores para plotar o grafico
w1, w2, w3 = perceptron.weight.data.numpy()[0]
b = perceptron.bias.data[0]
```

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # Importa para permitir plots 3D


# Função para plotar o plano 3D
def plot3d(perceptron_model):
    w1, w2, w3 = perceptron.weight.data.numpy()[0]
    b = perceptron.bias.data[0]

    # Definimos um intervalo para x1 e x2
    x1_range = np.linspace(-1, 1, 10)
    x2_range = np.linspace(-1, 1, 10)
    X1, X2 = np.meshgrid(x1_range, x2_range)

    # Calculando X3 a partir da equação do plano
    # x3 = (-b - w1*x1 - w2*x2) / w3
    X3 = (b - w1 * X1 - w2 * X2) / w3

    # Configurando a figura para o plot 3D
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(projection='3d')
    ax.plot_surface(X1, X2, X3, cmap='plasma')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('X3')
    ax.set_title('Plano de Decisão do Perceptron')


    ax.view_init(azim=180)


plot3d(perceptron)
```
