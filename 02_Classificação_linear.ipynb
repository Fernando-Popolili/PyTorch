#**Perceptron - Classificação Linear**

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Equação Linear

ax + by + c = 0



```python
# Valores na nossa função
a = -1
b = 4
c = 0.4

# ax + by + c -> Equação da classificação linear
# y = (-a*x -c)/b -> Isolando o Y
def plotline(a,b,c):
  x = np.linspace(-2, 4, 50) #Definindo valores para o X, (50 valores de -2 à 4)
  y = (-a*x - c)/b #Definindo a esquação de classificação linear
  plt.axhline(0, color='black', linestyle='--')
  plt.axvline(0, color='black', linestyle='--')
  plt.plot(x, y)
  plt.grid(True)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Classificação Linear')


plotline(a, b, c)
```

### Funcionamento de uma classificação linear
<p> Ao solucionar uma equação linear:</p>


*   Se o resultado for igual à 0 é pq está em cima da reta (ax + by + c = 0)
*   Se der negativo está abaixo
*   Se der positivo está acima


```python
#Ao solucionar a equação, se der 0 é pq está em cima da reta (ax + by + c = 0)
# Se der negativo está abaixo
# Se der positivo está acima
p1 = (2, 0.4)
p2 = (1, 0.6)
p3 = (3, 0.2)

ret1 = a*p1[0] + b*p1[1] + c
print("%.2f" % ret1)

ret2 = a*p2[0] + b*p2[1] + c
print("%.2f" % ret2)

ret3 = a*p3[0] + b*p3[1] + c
print("%.2f" % ret3)
```

```python
#Plotando os pontos no grafico

plotline(a, b, c)
plt.plot(p1[0], p1[1], color = 'b', marker = 'o')
plt.plot(p2[0], p2[1], color = 'r', marker = 'o')
plt.plot(p3[0], p3[1], color = 'g', marker = 'o')
```

### Modelo linear

<p>ax + by + c</p>

"a" e "b" equivalem a pesos,  c equivela ao bias(viés) , "X" e "Y" chamamos de dimensões (Entradas)
